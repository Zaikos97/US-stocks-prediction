```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
players=pd.read_csv(r'C:\Users\Admin\Desktop\Bootcamp\athanasios-zaikos-exercises\NBA Analysis Project\Cleaning\clean.csv')
pd.set_option('display.max_columns', None)
```

```python
players.head(5)
```

# Finding the 10 players with the highest BPM

```python
best_players=players.copy()
best_players=best_players[best_players['MP']>=20.0]
best_players=best_players.sort_values('BPM',ascending=False).head(10).reset_index(drop=True)
best_players
```

```python
name = best_players['Player']
price = best_players['BPM']
fig, ax = plt.subplots(figsize =(16, 9))
ax.barh(name, price)
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(True)
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
ax.invert_yaxis()
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
ax.set_title('BPM',
             loc ='center', )
fig.text(0.9, 0.15, 'Jeeteshgavande30', fontsize = 12,
         color ='grey', ha ='right', va ='bottom',
         alpha = 0.7)
plt.show()
```

# Finding the 30th value for each category

```python
stats=['PTS','TRB','AST','STL','BLK']
avg={}
for x in stats:
    players=players.sort_values(x,ascending=False).reset_index(drop=True)
    avg[x]=players[x][29]
avg
```

```python
plt.bar(range(len(avg)), list(avg.values()), align='center')
plt.xticks(range(len(avg)), list(avg.keys()))
for y in avg.keys():
    plt.text(list(avg).index(y),avg[y],avg[y],ha='center',va='bottom')
plt.show()
```

# Finding if top players are best on a specific attribute or they are allrounders

```python
best_players['All_around'] = 'e'
counter=0
for y in range(0,10):
    counter=0
    for x in avg.items():
        if best_players[x[0]][y]>=x[1]:
            counter+=1
        else:
            continue
    if counter >=3:
        best_players['All_around'][y]='Yes'
    else:
        best_players['All_around'][y]='No'
best_players
```

```python
pie=best_players.groupby('All_around').count()
pie
```

```python
my_data = pie['Player']
my_labels = 'No','Yes'
my_colors = ['lightblue','lightsteelblue','silver']
my_explode = (0, 0.1)
plt.pie(my_data,labels=my_labels,autopct='%1.1f%%',startangle=15, shadow = True, colors=my_colors, explode=my_explode)
plt.title('All_around')
plt.axis('equal')
plt.show()
```

# Specific comparison for Lebron James

```python
LJ=[best_players['PTS'][3],best_players['TRB'][3],best_players['AST'][3],best_players['STL'][3],best_players['BLK'][3]]
```

```python
labels = list(avg.keys())
Lebron_James = LJ
top30 = list(avg.values())

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, Lebron_James, width, label='Lebron James')
rects2 = ax.bar(x + width/2, top30, width, label='top30')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores per Category')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
for x in LJ:
    plt.text(LJ.index(x),x,x, ha="right", va="bottom")
for y in avg.keys():
    plt.text(list(avg).index(y),avg[y],avg[y],ha='left',va='bottom')

fig.tight_layout()

plt.show()
```
